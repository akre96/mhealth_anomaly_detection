```python
from typing import Tuple

def reformat_data(
    data:pd. DataFrame
) -> Tuple[pd.DataFrame, pd.DataFrame]:
    # -1 id seems to be a testing account, has many ~25 eureka ids
    reform_data = data[data.study_id != -1].copy()
    reform_data['subject_id'] = reform_data['study_id']

    # Study appears broken in to ~3 distinct sections separated artificially by 10,000 days    
    reform_data.loc[reform_data.day > 2.017*(10**7), 'stage'] = 3
    reform_data.loc[reform_data.day < 2.017*(10**7), 'stage'] = 2
    reform_data.loc[reform_data.day < 2.016*(10**7), 'stage'] = 1
    reform_data.loc[reform_data.day < 2*(10**7), 'stage'] = 0

    first_days = pd.DataFrame(reform_data.groupby(['subject_id', 'stage'])['day'].min()).reset_index().rename(columns={'day': 'first_day'})
    reform_data = reform_data.merge(first_days, validate='m:1')
    reform_data['study_day'] = reform_data['day'] - reform_data['first_day']

    # merge 70 day gaps
    subject_changes = {
        'subject_id': [],
        'stage': [],
        'eureka_id': [],
        'day': []
    }
    gap_70_day_merged = []
    for (sid, stage, eid), subject_df in reform_data.groupby(['subject_id', 'stage', 'eureka_id']):
        for i, val in enumerate(subject_df.study_day.diff()):
            if (val >= 70) and (val < 75):
                subject_df['study_day'].iloc[i:] = subject_df.iloc[i:]['study_day'] - 70
                subject_changes['subject_id'].append(sid)
                subject_changes['stage'].append(stage)
                subject_changes['eureka_id'].append(eid)
                subject_changes['day'].append(subject_df.iloc[i]['day'])
                
        gap_70_day_merged.append(subject_df)
    
    reform_data = pd.concat(gap_70_day_merged)

    # Stage 0 days seem like a bug
    return pd.DataFrame(subject_changes), reform_data[reform_data.stage > 0]

subject_changes, reform_cc_data = reformat_data(cc_data)
reform_cc_data
```
